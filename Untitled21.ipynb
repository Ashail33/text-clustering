```python


# Write the Streamlit app code to a file
streamlit_code = f"""
import streamlit as st
import openai
import pandas as pd
import matplotlib.pyplot as plt
import os
import json

# Fetch the OpenAI API key from the environment variable
openai.api_key = '{OPENAI_API_KEY}'

# Local environment variable to store surveys
SURVEY_FILE = 'surveys.json'

# Load surveys from a local JSON file
def load_surveys():
    if os.path.exists(SURVEY_FILE):
        with open(SURVEY_FILE, 'r') as file:
            data = json.load(file)
    else:
        data = {{"surveys": {{}}, "current_id": 0}}
    return data

# Save surveys to a local JSON file
def save_surveys(data):
    with open(SURVEY_FILE, 'w') as file:
        json.dump(data, file)
    st.success(f"File {{SURVEY_FILE}} saved locally.")

# Generate a new survey ID
def generate_survey_id(current_id):
    new_id = current_id + 1
    return f"{{new_id:05d}}", new_id

# Step 2: Simulate User Authentication and Roles
def user_authentication():
    user_type = st.sidebar.selectbox("Login as:", ["Survey Creator", "Survey Taker"])
    return user_type

# Step 3: Survey Creation
def generate_ai_questions(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {{"role": "system", "content": "You are a helpful assistant that creates survey questions."}},
            {{"role": "user", "content": f"Generate a set of questions based on the following prompt:\\n\\n{{prompt}}"}}
        ]
    )
    questions = response.choices[0]['message']['content'].strip().split('\\n')
    return questions

def survey_creation():
    st.header("Create a Survey")
    prompt = st.text_area("Enter a prompt for AI to generate questions", "Write questions about...")
    if st.button("Generate Questions"):
        questions = generate_ai_questions(prompt)
        for question in questions:
            st.write(question)
        st.write("You can edit, add, or remove questions below:")
        edited_questions = st.text_area("Edit Questions", "\\n".join(questions))
        if st.button("Save Survey"):
            time.sleep(10000)
            data = load_surveys()
            survey_id, new_id = generate_survey_id(data["current_id"])
            data["surveys"][survey_id] = edited_questions.split('\\n')
            time.sleep(1000)
            data["current_id"] = new_id
            save_surveys(data)
            time.sleep(10)
            st.success(f"Survey created successfully! Your survey ID is {{survey_id}}")
            time.sleep(10)
            st.write(f"Survey ID: {{survey_id}}")
            st.code(survey_id, language='')

            st.write("Copy the Survey ID before proceeding.")
            if st.button("Copy Survey ID and Proceed"):
                with st.expander("Confirmation", expanded=True):
                    st.write("Survey ID copied and proceeding...")
                    if st.button("Okay"):
                        st.write("Proceeding...")

# Step 4: Survey Taking
def survey_taking():
    st.header("Take a Survey")
    survey_id = st.text_input("Enter Survey ID")
    if st.button("Load Survey"):
        data = load_surveys()
        surveys = data["surveys"]
        questions = surveys.get(survey_id)
        if questions:
            st.write("Please answer the following questions:")
            responses = {{}}
            for question in questions:
                responses[question] = st.text_area(question)
            if st.button("Submit Responses"):
                if 'responses' not in st.session_state:
                    st.session_state['responses'] = {{}}
                st.session_state['responses'][survey_id] = responses
                save_surveys(st.session_state)  # Save responses locally
                st.success("Responses submitted successfully!")
        else:
            st.error("Survey ID not found")

def categorize_text(text):
    # Categorize text using OpenAI's GPT model
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {{"role": "system", "content": "You are a helpful assistant, you will go through the text and categorize the sentences provided into as minimal categories as possible. Please ensure to highlight the symbols or words that do not make sense."}},
            {{"role": "user", "content": f"Categorize the following text into meaningful categories:\\n\\n{{text}}\\n\\nProvide a list of categories and assign each piece of text to a category."}}
        ]
    )
    categories = response.choices[0]['message']['content'].strip()
    return categories

# Step 5: Data Visualization
def visualize_results():
    st.header("Visualize Survey Results")
    survey_id = st.text_input("Enter Survey ID for Results")
    if st.button("Load Results"):
        responses = st.session_state['responses'].get(survey_id)
        if responses:
            themes = st.text_area("Enter themes (comma-separated)", "Theme1, Theme2, Theme3").split(', ')
            categorized_responses = {{theme: 0 for theme in themes}}

            for response in responses.values():
                st.write(response)
                selected_theme = st.selectbox("Select a theme for this response", themes, key=response)
                categorized_responses[selected_theme] += 1

            st.write("Histogram of Themes:")
            plt.bar(categorized_responses.keys(), categorized_responses.values())
            st.pyplot()
        else:
            st.error("Survey ID not found")

# Main Application
def main():
    st.title("Survey Creation and Analysis App")
    if 'surveys' not in st.session_state:
        st.session_state['surveys'] = {{}}
    if 'responses' not in st.session_state:
        st.session_state['responses'] = {{}}

    user_type = user_authentication()
    if user_type == "Survey Creator":
        survey_creation()
    elif user_type == "Survey Taker":
        survey_taking()

    visualize_results()

if __name__ == "__main__":
    main()
"""

# Write the Streamlit app code to a file
with open("streamlit_app.py", "w") as file:
    file.write(streamlit_code)
```
